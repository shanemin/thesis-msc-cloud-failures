```python
import time
import glob as glob
from zipfile import ZipFile
from bs4 import BeautifulSoup
import dateutil.parser as dparser
from datetime import datetime
import pandas as pd
import multiprocessing as mp
import numpy as np
import io
import json
import sys
```

```python
# root_dir = '/media/shane/cloud-availability-sacheen-2020-05-11/cloud-amazon-web-services' # one directory
# root_dir = '/media/shane/cloud-availability-sacheen-2020-05-11/cloud-amazon-web-services/2020/20204' #subset of above
root_dir = '/media/shane/cloud-availability-sacheen-2020-05-11/cloud-amazon-web-services*' # all directories
```

```python
def read_file(file):
    with ZipFile(file, "r") as z:
        for filename in z.namelist(): 
            if 'json' in filename: 
                with z.open(filename) as f:  
                    data = f.read()
                    d = json.loads(data) 
                    ret = []
                    for v in d.values():
                        for v2 in v:
                            ret.append([v2])
                    return ret
```

```python
files = glob.glob(root_dir + '/**/*.zip', recursive=True)
pool = mp.Pool(mp.cpu_count())
start_time = time.time()
try:
    results = pool.map(read_file, [file for file in files])
except:
    print("Unexpected error:", sys.exc_info()[0])
    raise
pool.close()
end_time = time.time()
```

```python
unique = set()
counter = 0
try:
    for i in range (0, len(results)):
        for j in range(0, len(results[i])):
            counter += 1
            unique.add(str(results[i][j]).replace('[{','').replace('}]','').replace("\'",''))
except:
    pass
```

```python
type(results), counter, len(unique)
```

```python
df_unique = pd.DataFrame(columns=['service_name', 'summary', 'date', 'status', 'service'])
for entry in unique:
    arr = []
    try:
        arr.append(entry.split(':')[1].split(',')[0].strip())
        arr.append(entry.split(':')[2].split(',')[0].strip())
        ts = entry.split(':')[3].split(',')[0].strip()
        arr.append(datetime.utcfromtimestamp(int(ts)).strftime('%Y-%m-%d %H:%M:%S'))
        arr.append(entry.split(':')[4].split(',')[0].strip())
        arr.append(entry.split(':')[-1].split(',')[0].strip())
        df_unique.loc[len(df_unique)] = arr
    except:
        print(entry, '\n')
```
